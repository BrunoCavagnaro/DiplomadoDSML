## Tokenization

```
import en_core_web_sm
import spacy

# Import spaCy and load the language library
#import spacy
nlp = en_core_web_sm.load()

# Create a Doc object
doc = nlp(movie_review)

# Print each token separately
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```
#TOTAL DE ORACIONES EN LA REVIEW

sends = [send for send in doc.sents]
a = len(sends) 
```

```
#Imprimir cada oración:
for i in range(a):
  print('#'+str(i +1) + ' Sentence: ' + str(sends[i]))
```

## Visualizadores

```
#VISUALIZAR LOS ELEMENTOS DE UNA ORACIÓN RANDOM
from spacy import displacy
from random import randint
r = randint(0,a)

displacy.render(sends[r], style='ent', jupyter=True)
```

## Stemming


```
#ELIMINAR LOS SIGNOS DE PUNTUACIÓN PARA HACER EL STEMMING:
tokens = []
for token in doc:
    if (token.is_punct == False):
      tokens.append(token.text)
```

```
print(tokens)
```

```
import nltk
from nltk.stem.porter import *
p_stemmer = PorterStemmer()

for token in tokens:
  print(token+' --> '+p_stemmer.stem(token)) 
```

## Lemmatization

```
def show_lemmas(text):
    for token in text:
        print(f'{token.text:{12}} {token.pos_:{6}}  {token.lemma_}')

show_lemmas(doc)
```

```
#ELIMINAR STOPWORDS:
from spacy.lang.en.stop_words import STOP_WORDS

# Create list of word tokens after removing stopwords
filtered_sentence =[] 

for word in tokens:
    if (nlp.vocab[word].is_stop == True):
        filtered_sentence.append(word) 

print(filtered_sentence)  
remove = [x for x in tokens if x not in filtered_sentence]
print(remove)
print(len(filtered_sentence))
remove_duplicate = list(dict.fromkeys(remove))
print(remove_duplicate)
print(len(remove_duplicate))
```

## Predecir Sentiment Analysis usando lematizador con un Count Vectorizer

```
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer 
import nltk
nltk.download('punkt')
nltk.download('wordnet')
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, articles):
        return [self.wnl.lemmatize(t) for t in word_tokenize(articles)]

tf_vectorizer = CountVectorizer(tokenizer=LemmaTokenizer(),
                                strip_accents = 'unicode', # works 
                                stop_words = 'english', # works
                                lowercase = True, # works
                                max_df = 1.0, # works
                                min_df = 1) # works
```

```
import numpy as np
import pandas as pd

df = pd.read_csv('amazon_review.csv', sep=';')
df
```

```
X = df['text']  # this time we want to look at the text
Y = df['rating']
df_vect = tf_vectorizer.fit_transform(X)
df_vect.shape
```

```
#NORMALIZAR VECTORIZER
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()

X_train_tfidf = tfidf_transformer.fit_transform(df_vect)
X_train_tfidf.shape
```

```
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train_tfidf,Y)
```

```
# Form a prediction set
predictions = clf.predict(X_train_tfidf)
predictions
```

```
from sklearn import metrics
print(metrics.confusion_matrix(Y,predictions))
```

## Sentiment Analysis - Vader Lexicon

```
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```
scores = []
for i in range(len(df)):
  scores.append(sid.polarity_scores(df.text[i]))
```

```
df['scores'] = scores
df
```

```
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
df['comp_score'] = df['compound'].apply(lambda c: 1 if c >=0 else 0)
df
```

```
#CONFUSION MATRIX:
print(metrics.confusion_matrix(df.rating,df.comp_score))
```
